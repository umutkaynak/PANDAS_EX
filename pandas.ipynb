İLERİ TOPLULAŞTIRMA İŞLEMLERİ

```python
#AGGRIGATE  kullanımı
import pandas as pd 
df = pd.DataFrame({'gruplar': ['A', 'B', 'C', 'A', 'B', 'C'],
'degisken1': [10,23,33,22,11,99],
'degisken2': [100,253,333,262,111,969]},
columns = ['gruplar', 'degisken1', 'degisken2'])

df
 
```

```python
df.groupby("gruplar")
```

```python
df.groupby("gruplar").mean()
```

```python
#aggregate kullanımı içerisinde yapılacak olan işlemleri dışardan veya farklı bir kütüphaneden almayacaksak 
#""(tırnak işareti içerisidne kullanamda veya kullanamamada bir sıkıntı yaşamayız)
#ama dışardan alınan veya kütüphandeden alınan verilerde veya fonki ""(tırnak işareti ile kullanırsak) hata alırız

df.groupby("gruplar").aggregate(["min",np.median,max])

```

```python
#biribirinden bağımsız olarak aynı verisetindeki verilerden istatiksel hesaplamalar yapmak
df.groupby("gruplar").aggregate({"degisken1": "min", "degisken2": max})
```

```python
#FİLTER KOMUTU
#karılşık veri setinde kullanıcı derleyerek hazırlanan kod. 
df = pd.DataFrame({'gruplar': ['A', 'B', 'C', 'A', 'B', 'C'],
'degisken1': [10,23,33,22,11,99],
'degisken2': [100,253,333,262,111,969]},
columns = ['gruplar', 'degisken1', 'degisken2'])

df
```

```python
def filter_func(x):
    return x["degisken1"].std()>9
```

```python
df.groupby("gruplar").filter(filter_func)
```

```python
#transform yapısı
#veri setinin değişknelerine dönüştürme iişemi yapıldığında kullanılır. yaptığımız dönüştürme işleminide kendimiz tanımlıyoruz
df["degisken1"]*9
```

```python
df.transform(lambda x : x-x.mean)
```

```python
df_a=df.iloc[: , 1:3]
df_a
```

```python
#x den xin ortalama değerlerini çıkarttık
df_a.transform(lambda x: x-x.mean())
```

```python
#APPLY fonk kullanımı:
#df değişken üzerinde gezinme yeteneği olan ve aggregate(toplulaştırma) gibi olan fonksiyonlardır

df = pd.DataFrame({
'degisken1': [10,23,33,22,11,99],
'degisken2': [100,253,333,262,111,969]},
columns = ['degisken1', 'degisken2'])

df
```

```python
df.apply(np.sum)
```

PİVOT TABLOLAR


```python
import pandas as pd
import seaborn as sns
titanic=sns.load_dataset("titanic")
titanic.head()
```

```python
titanic.groupby("sex")[["survived"]].mean()
```

```python
#unstack kavramı satır bazlı olan index yapısını sütun abzlı yapar ve daha rahat okunur
titanic.groupby(["sex","class"])[["survived"]].aggregate("mean").unstack()
```

```python
#pivot ile table
titanic.pivot_table("survived", index="sex", columns="class")
```

```python
# çok boyutlu yapma 
titanic.age.head()
```

```python
age= pd.cut(titanic["age"], [0,18,90])
age.head(10)
```

```python
titanic.pivot_table("survived",["sex",age], "class")
```
